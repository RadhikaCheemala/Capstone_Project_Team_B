```python
matches_model['winner'] = matches_model.apply(lambda row: 'draw' if row['home_team_goal_count'] == row['away_team_goal_count'] 
                        else 'home' if row['home_team_goal_count'] > row['away_team_goal_count'] 
                        else 'away', axis=1)
```

```python
grouped_df_home_analysis = matches_model.groupby('home_team_name')[['home_team_shots_on_target', 
                                                                    'home_team_shots_off_target', 
                                                                    'home_team_fouls',
                                                                    'home_team_possession']].median().reset_index()
```

```python
grouped_df_home_analysis
```

```python
grouped_df_away_analysis = matches_model.groupby('away_team_name')[['away_team_shots_on_target', 
                                                               'away_team_fouls',
                                                               'away_team_shots_off_target', 
                                                               'away_team_possession']].median().reset_index()
```

```python
grouped_df_away_analysis
```

```python

merged_df = matches_model.merge(grouped_df_home_analysis, on='home_team_name', suffixes=('', '_mean'))

# Fill NaN values in matches_model with values from grouped_df_home_analysis
columns_to_fill = ['home_team_shots_on_target', 'home_team_shots_off_target', 'home_team_fouls', 'home_team_possession']

for col in columns_to_fill:
    merged_df[col] = merged_df[col].fillna(merged_df[f"{col}_mean"])

# Drop the extra columns created during the merge
merged_df = merged_df.drop(columns=[f"{col}_mean" for col in columns_to_fill])
```

```python
matches_model = merged_df
```

```python
matches_model.isnull().sum()
```

```python

merged_df = matches_model.merge(grouped_df_away_analysis, on='away_team_name', suffixes=('', '_mean'))

# Fill NaN values in matches_model with values from grouped_df_home_analysis
columns_to_fill = ['away_team_shots_on_target', 'away_team_shots_off_target', 'away_team_fouls', 'away_team_possession']

for col in columns_to_fill:
    merged_df[col] = merged_df[col].fillna(merged_df[f"{col}_mean"])

# Drop the extra columns created during the merge
merged_df = merged_df.drop(columns=[f"{col}_mean" for col in columns_to_fill])
```
